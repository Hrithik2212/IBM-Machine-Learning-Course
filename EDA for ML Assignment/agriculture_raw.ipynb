### 3) Actions taken for data cleaning and feature engineering



##### 3.1)Drop Out Missing Values 

```python
# We are dropping out rows with missing data 

data = data.dropna()
data.info()
```

```python
data.head()
```

##### 3.2)
##### The data set has many columns with numericals but the default dtype for them is object 
#####     Thus we will remove the percentage sign and covert their dtype to float64 

```python
# List for columns with dtype as object other than Month column
col_objt=[]
for i in data.columns:
    if data[i].dtype==object:
        col_objt.append(i)
col_objt.remove('Month')
print(col_objt)
```

```python
data=data.replace('-', '0')
```

```python
# We convert obj data types of cost price and cost % price to float64
for i in col_objt:
    data[i]=data[i].str.replace('%', '')
    data[i]=data[i].str.replace(',', '')
    data[i]=data[i].astype('float')
```

```python
data.describe()
```

```python
data.head()
```

```python
lst_col=list(data.columns)
lst_col_price=[]
for i in range (3,len(lst_col),2):
    lst_col_price.append(lst_col[i])
lst_col_price.append('Coarse wool Price')

```

```python
plt.figure(figsize=(15,15))
for col in lst_col_price:
    plt.plot(data['Month'],data[col],label=col.replace("Price",""))
plt.xlabel("Month")
plt.ylabel("Price in cents")
plt.title("Prices of Agriculture Product Vs Month")
plt.legend()
```

```python
lst_col=list(data.columns)
lst_col_percent_price=[]
for i in range (2,len(lst_col),2):
    lst_col_percent_price.append(lst_col[i])
```

```python
plt.figure(figsize=(15,15))
for col in lst_col_percent_price:
    plt.plot(data['Month'],data[col],label=col.replace("price % change",""))
plt.xlabel("Month")
plt.ylabel("Price in cents")
plt.title("Prices of Agriculture Product Vs Month")
plt.legend()
```
